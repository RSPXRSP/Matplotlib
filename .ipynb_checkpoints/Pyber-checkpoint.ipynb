```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

city_file = '/Users/rspxrsp/Documents/TECMC201902DATA2/02-Homework/05-Matplotlib/Instructions/Pyber/data/city_data.csv'
ride_file = '/Users/rspxrsp/Documents/TECMC201902DATA2/02-Homework/05-Matplotlib/Instructions/Pyber/data/ride_data.csv'
```

```python
city_data = pd.read_csv(city_file)
city_data = city_data.drop_duplicates('city')
ride_data = pd.read_csv(ride_file)
pyber_data = pd.merge(city_data, ride_data, on="city")

mean_fare = pyber_data.groupby("city")["fare"].mean()
total_rides = pyber_data.groupby("city")["ride_id"].count()
total_drivers = pyber_data.groupby("city")["driver_count"].sum()

city_type = city_data.set_index("city")["type"]
city_type.value_counts()

pyber_cities = pd.DataFrame({"Average Fare per City": mean_fare,
                 "Number of Rides": total_rides,
                 "Number of Drivers": total_drivers,
                 "City Type": city_type})

urban = pyber_cities.loc[pyber_cities["City Type"] == "Urban"]
suburban = pyber_cities.loc[pyber_cities["City Type"] == "Suburban"]
rural = pyber_cities.loc[pyber_cities["City Type"] == "Rural"]

pyber_cities.head()
```

## Bubble Plot of Ride Sharing Data

```python
plt.scatter(urban["Number of Rides"], urban["Average Fare per City"], color = "gold", edgecolors="black", s = urban["Number of Drivers"]*0.5, label = "Urban", alpha = 0.5, linewidth = 1.5)
plt.scatter(suburban["Number of Rides"], suburban["Average Fare per City"], color = "lightskyblue", edgecolors ="black", s = suburban["Number of Drivers"]*1, label = "Suburban", alpha = 0.5, linewidth = 1.5)
plt.scatter(rural["Number of Rides"], rural["Average Fare per City"], color = "lightcoral", edgecolors = "black", s = rural["Number of Drivers"]*2, label = "Rural", alpha = 0.5, linewidth = 1.5)
plt.title("Average Fare per City versus Number of Rides")
plt.xlabel("Number of Rides")
plt.ylabel("Average Fare Per City")
plt.show()
```

## Total Fares by City Type

```python
city_fare_pie = pyber_data.groupby(["type"])["fare"].sum()
city_fare_pie
city_fare_pie.index
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0,0, 0.1)
plt.pie(city_fare_pie, labels = city_fare_pie.index, autopct = "%1.2f%%", colors = colors, explode = explode, shadow = True, startangle = 270)
plt.axis("equal")
plt.legend(loc = "upper right")
plt.show()
```

## Total Rides by City Type

```python
city_rides_pie = pyber_data.groupby(["type"])["ride_id"].count()
city_rides_pie
city_rides_pie.index
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0.5,0,0)
plt.pie(city_rides_pie, labels = city_fare_pie.index, autopct = "%1.2f%%", colors = colors, explode = explode, shadow = True)
plt.axis("equal")
plt.legend(loc = "upper right")
plt.show()
```

## Total Drivers by City Type

```python
city_drivers_pie = pyber_data.groupby(["type"])["driver_count"].sum()
city_drivers_pie
city_drivers_pie.index
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0.2,0.2,0)
plt.pie(city_drivers_pie, labels = city_fare_pie.index, autopct = "%1.2f%%", colors = colors, explode = explode, shadow = True)
plt.axis("equal")
plt.legend(loc = "upper right")
plt.show()
```
